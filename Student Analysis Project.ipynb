```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
analysis_df = pd.read_csv("student_info.csv")
print(analysis_df)
```

```python
print("To print the first 5 rows of the student details")
print(analysis_df.head())
```

```python
print("What is the gender breakdown of the students?")
print(analysis_df['gender'].value_counts())
```

```python
print("What is the age distribution of the students?")
print(analysis_df['age'].value_counts())
```

```python
# 1. Count how many students are in each grade level
grade_counts = analysis_df['grade_level'].value_counts()
print(grade_counts)
```

```python
# 2. Plot it as a clean bar chart
grade_counts.plot(kind='bar', color='violet', edgecolor='black')

# 3. Add simple labels to understand well
plt.title('Number of Students per Grade Level')
plt.xlabel('Grade Level')
plt.ylabel('Number of Students')

# 4. To display it
plt.xticks(rotation=0) # Keeps the numbers at the bottom standing straight up
plt.tight_layout()
plt.show()
```
